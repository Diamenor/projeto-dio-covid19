### Padronização do nome das colunas

ESCREVER ALGO AQUI

```python
import re

#substituir "/","|" e " " (barra,pipe-line e espaço) por "" (caracter vazio)
def corrige_colunas(col_name):
    return re.sub(r"[/| ]", "", col_name).lower()
```

```python
df.columns = [corrige_colunas(col) for col in df.columns]
```
